## Feature Engineering and Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix,  ConfusionMatrixDisplay
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
df.columns
```

```python
# Dropping irrelevant columns
df_model = df.drop(["step", "nameOrig", "nameDest", "isFlaggedFraud"], axis=1)
```

```python
df_model.info()
```

```python
# Defining column type
categorical = ["type"]
numeric = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
```

```python
# Defining the features and target variable
y = df_model['isFraud']
X = df_model.drop(['isFraud'], axis=1)
```

```python
# Train Test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
```

```python
# Creating a preprocessing pipeline
# Dropping first category of the categorical column to avoid perfect multicolinearity (type_1 = 1 - type2 + type_3)
preprocessor = ColumnTransformer(
    transformers = [
        ('num', StandardScaler(), numeric),
        ('cat', OneHotEncoder(drop='first'), categorical)
    ],
    remainder='drop'
)
```

```python
# Generating the final pipeline
# class_weight = "balanced" give more weights to errors for minority class while optimizing it
pipeline = Pipeline(
    [
        ("prep", preprocessor),
        ("clf", LogisticRegression(class_weight="balanced", max_iter=1000))
    ]
)
```

```python
# fitting the model
pipeline.fit(X_train, y_train)
```

```python
# Generating the predictions
y_pred = pipeline.predict(X_test)
```

```python
# Generaring the classification report
print(classification_report(y_test, y_pred))
```

```python
# Generating the confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Negative', 'Positive'])
disp.plot(cmap='Blues', values_format='d')
plt.title("Confusion Matrix")
plt.show()
```

```python
# Exporting the pipeline as pkl file
import joblib

joblib.dump(pipeline, "fraud_detection_pipeline.pkl")
```
